```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Musicology_Dataset.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

df["Time Period"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    )

plt.title("Time Period Focus")
plt.show()
```

```python
df['Subfield_split'] = df['Subfield'].astype(str).str.split(';')
df[['Subfield', 'Subfield_split']].head()
```

```python
df_exploded = df.explode('Subfield_split')
df_exploded['Subfield_split'] = df_exploded['Subfield_split'].str.strip().str.lower()
df_exploded.head(20)
```

```python
subfield_counts = df_exploded['Subfield_split'].value_counts().head(20)

subfield_counts.sort_values().plot(
    kind='barh',
    color='mediumpurple',
    edgecolor='black'
)

plt.title('Top 10 Subfields Mentioned')
plt.xlabel('Number of Articles')
plt.ylabel('Subfield')
plt.tight_layout()
plt.show()
```

```python
df['Accessed_split'] = df['How Accessed?'].astype(str).str.split(';')
df[['How Accessed?', 'Accessed_split']].head()
```

```python
df['Accessed_split'] = df['How Accessed?'].astype(str).str.split(';')

df_access_exploded = df.explode('Accessed_split').copy()
df_access_exploded['Accessed_split'] = (
    df_access_exploded['Accessed_split'].astype(str).str.strip().str.lower()
)
df_access_exploded[['How Accessed?', 'Accessed_split']].head()
```

```python
import matplotlib.pyplot as plt

counts = df_access_exploded['Accessed_split'].value_counts().head(10)

counts.sort_values().plot(kind='barh', color='teal', edgecolor='black')
plt.title('Top 10 Access Types')
plt.xlabel('Number of Articles')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
