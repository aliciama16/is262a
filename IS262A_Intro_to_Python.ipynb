# Loops

for loops let us iterate through a definite set of objects.

In each iteration through the for loop, Python will:

* Extract one element from the dataset
* Execute the body of the for loop using the item bound to the element
* Go back to the first step
* Keep iterating through the loop until reaching the end of the dataset



```python
types = ["casettes", "CDs", "films", "Vinyls"]

for item in types:
  print("Processing:", item)
```

# Transformation

In Python, a transformation means changing data from one form into another. You can think of it as reformatting or standardizing information so it's easier to work with or compare.

Here is an example of a string transformation... capitalization!

```python
types = ["casettes", "CDs", "films", "Vinyls"]

for item in types:
  print("Processing:", item.upper())
```

```python
types = ["casettes", "CDs", "films", "Vinyls"]

for item in types:
  print("Processing:", item.lower())
```

Transformations can also look like:
* removing extra spaces
* converting formats
* combining columns
* splitting values

Sometimes a cell holds TOO much information (like a list of items seperated by commas or semicolons. Splitting these cells means breaking that text apart into smaller, individual pieces. Imagine a smoothie. Splitting is like seperating it back to its ingredients so we can see the individual parts inside.

```python
data = "interviews; surveys; photographs; letters; census reports; newspaper articles"

split_data = data.split(";") #this tells python to SPLIT the data every time it sees a semicolon
print(split_data)
```
